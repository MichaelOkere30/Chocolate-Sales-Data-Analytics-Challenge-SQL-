```python
import pandas as pd
from sqlalchemy import create_engine
from pathlib import Path
from db_utils import psql_insert_copy
```

```python
cnxn_string = ("postgresql+psycopg2://{username}"
               "@{host}:{port}/{database}")
```

```python
engine = create_engine(cnxn_string.format(
    username = "postgres",
    host = "localhost",
    port = 5432,
    database = "choco_sales_db"))
```

```python
with engine.connect() as conn:
    print("Connected successfully!")
```

```python
csv_folder = Path(r"C:\Users\user\Desktop\Chocolate_Sales_Dataset_2023_-_2024")
```

```python
for csv_file in csv_folder.glob("*.csv"):
    table_name = csv_file.stem.lower()

    print(f"Loading {csv_file.name}...")

    df = pd.read_csv(csv_file)

    df.to_sql(
        table_name,
        engine,
        if_exists="replace",
        index=False,
        method=psql_insert_copy
    )

    print(f"Created table: {table_name}")

print("All CSV files imported.")
```
